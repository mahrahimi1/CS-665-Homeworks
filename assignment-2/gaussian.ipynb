```python
import matplotlib.pyplot as plt
import random
import math
```

```python
def unit_normal(d):
    s = (1.0 / d) ** 0.5
    return list(random.normalvariate(0, s) for _ in range(d))


def get_euclidean_distance(instance1 , instance2):
    sum = 0.0
    for i in range(len(instance1)):
        sum = sum + math.pow((instance1[i] - instance2[i]) , 2)
    return math.sqrt(sum)
    

def get_vector_length(vector):
    sum = 0.0
    for i in range(len(vector)):
        sum = sum + math.pow( vector[i] , 2)
    return math.sqrt(sum)


def inner_product(v1, v2):
    return sum((a * b) for (a, b) in zip(v1, v2))
```

```python
# 2-D samples
X = []
for i in range(500):
    X.append(unit_normal(2))

# 100-D samples
Y = []
for i in range(500):
    Y.append(unit_normal(100))
```

```python
X_lengths = []
for i in range(len(X)):
    X_lengths.append(get_vector_length(X[i]))
    
plt.hist(X_lengths , bins='auto') 
plt.title("Histogram of the lengths of the 2-D samples")
plt.show()

Y_lenghts = []
for i in range(len(Y)):
    Y_lenghts.append(get_vector_length(Y[i]))
    
plt.hist(Y_lenghts , bins='auto') 
plt.title("Histogram of the lengths of the 100-D samples")
plt.show()
```

```python
X_Pairs = []
for i in range(len(X)):
    for j in range(i+1 , len(X)):
        X_Pairs.append(get_euclidean_distance(X[i] , X[j]))

        
plt.hist(X_Pairs , bins='auto')
plt.title("Histogram of the distances between all pairs of the 2-D samples")
plt.show()

Y_Pairs = []
for i in range(len(Y)):
    for j in range(i+1 , len(Y)):        
        Y_Pairs.append(get_euclidean_distance(Y[i] , Y[j]))

plt.hist(Y_Pairs , bins='auto')
plt.title("Histogram of the distances between all pairs of the 100-D samples")
plt.show()
```

```python
X_pairs_innprd = []
for i in range(len(X)):
    for j in range(i+1 , len(X)):
        X_pairs_innprd.append(inner_product(X[i], X[j]))
        
plt.hist(X_pairs_innprd , bins='auto')
plt.title("Histogram of the inner products between all pairs of the 2-D samples")
plt.show()

Y_pairs_innprd = []
for i in range(len(Y)):
    for j in range(i+1 , len(Y)):
        Y_pairs_innprd.append(inner_product(Y[i] , Y[j]))

plt.hist(Y_pairs_innprd , bins='auto')
plt.title("Histogram of the inner products between all pairs of the 100-D samples")
plt.show()
```
